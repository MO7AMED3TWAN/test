# Importing The Libraries

```python
import warnings
warnings.filterwarnings('ignore')

## Basic Packages FOr Data Preprocessing
#FOR Data Manipulation
import pandas as pd
import numpy as np

#FOR Visualization
import matplotlib.pyplot as plt
import seaborn as sns

#Modeling Packages
from sklearn.model_selection import train_test_split #for spliting The data
from sklearn.preprocessing import StandardScaler #for Normalization
from sklearn.linear_model import LinearRegression #for The Model we need for regression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error #for Evaluation
```

# Step 1 - Gathering The Data

```python
df = pd.read_csv("/kaggle/input/students-exam-scores/Expanded_data_with_more_features.csv")
```

# Step 3 - Data Wrangling

## Handle Outlaiers

```python
for i in columns:
    print(f'THe Feature of {i} : \n{df[i].value_counts()}\n\n')
```

```python
df= df[df['ParentMaritalStatus'] != "widowed"]
```

```python
df.drop(columns=['Unnamed: 0', 'NrSiblings'], inplace= True)
```

### Handle Nulls

```python
# Checking null values
print(df.isnull().sum())
```

```python
df['EthnicGroup'].fillna(df['EthnicGroup'].mode()[0], inplace=True)
df['ParentEduc'].fillna(df['ParentEduc'].mode()[0], inplace=True)

df['TestPrep'].fillna(df['TestPrep'].mode()[0], inplace=True)
df['ParentMaritalStatus'].fillna(df['ParentMaritalStatus'].mode()[0], inplace=True)

df['PracticeSport'].fillna(df['PracticeSport'].mode()[0], inplace=True)
df['IsFirstChild'].fillna(df['IsFirstChild'].mode()[0], inplace=True)

df['TransportMeans'].fillna(df['TransportMeans'].mode()[0], inplace=True)
df['WklyStudyHours'].fillna(df['WklyStudyHours'].mode()[0], inplace=True)
```

```python
df.sample(3)
```

```python
# Checking null values
print(df.isnull().sum())
```

### Handle Object Features (Encoding)

```python
columns = ['Gender', 'EthnicGroup', 'ParentEduc', 'LunchType',
       'TestPrep', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild',
        'TransportMeans', 'WklyStudyHours']
```

```python
df.info()
```

```python
for i in columns:
    df[i], _ =  pd.factorize(df[i])
```

```python
df.info()
```

```python
df.sample(5)
```

### Final DATA

```python
# Standardization
scaler = StandardScaler()
```

```python
X_writing = df.drop(columns= 'WritingScore')
X_writing = scaler.fit_transform(X_writing)

X_reading = df.drop(columns= 'ReadingScore')
X_reading = scaler.fit_transform(X_reading)

X_math = df.drop(columns= 'MathScore')
X_math = scaler.fit_transform(X_math)

y_writing = df['WritingScore']
y_reading = df['ReadingScore']
y_math = df['MathScore']
```

# Step 4 - Modeling 

### Model Of Writing 

#### Split The Data 

```python
X_train_w, X_test_w, y_train_w, y_test_w = train_test_split (
    X_writing, y_writing,
    test_size= 0.2,
    shuffle = True
)
```

#### Training The Models

```python
Model_1_Of_Writting = LinearRegression()
```

```python
Model_1_Of_Writting.fit(X_train_w, y_train_w)
```

#### Evaluate The Models

##### For Train Phase

```python
# Make predictions
t_y_pred_w = Model_1_Of_Writting.predict(X_train_w)

# Evaluate metrics
t_r2_w = r2_score(y_train_w, t_y_pred_w)
t_mse_w = mean_squared_error(y_train_w, t_y_pred_w)
t_mae_w = mean_absolute_error(y_train_w, t_y_pred_w)

print("Model Evaluation Metrics:")
print(f"R² Score: {round(t_r2_w, 4) *100} %")
print(f"Mean Squared Error: {t_mse_w:.4f}")
print(f"Mean Absolute Error: {t_mae_w:.4f}")
```

##### For Test Phase

```python
# Make predictions
y_pred_w = Model_1_Of_Writting.predict(X_test_w)

# Evaluate metrics
r2_w = r2_score(y_test_w, y_pred_w)
mse_w = mean_squared_error(y_test_w, y_pred_w)
mae_w = mean_absolute_error(y_test_w, y_pred_w)

print("Model Evaluation Metrics:")
print(f"R² Score: {round(r2_w, 4) *100} %")
print(f"Mean Squared Error: {mse_w:.4f}")
print(f"Mean Absolute Error: {mae_w:.4f}")
```

### Saving The Model

```python
import pickle

with open('Model_1_Of_Writting.pkl', 'wb') as file:
    pickle.dump(Model_1_Of_Writting, file)

print("Model saved as 'Model_1_Of_Writting.pkl'")
```
